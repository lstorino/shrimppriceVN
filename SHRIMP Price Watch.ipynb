# Scrapping data from wbesite

## 1. Importing libraries essetial to scrapping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## 2. Pointing url to fetch data from

```python
url = 'https://tepbac.com/gia-thuy-san/cat/2-thuy-san/'
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
```

## 3. Locating tags where the data in relevant fetching into a dataframe

```python
markets = []
listings = soup.find_all('tr')
```

```python
markets = []
listings = soup.select('tr')
```

```python
print(listings)
```

```python
print(soup.prettify())
```

## 4. Fetching data from located tags in the HTML into a proper dataframe.

```python
for item in listings:
    market = {}
    
    market['specie'] = item.find('a', {'class':"h6"}).getText()
        
    
    
```

```python
for item in listings:
    market = {}

    market['specie'] = item.find('a', {'class':'h6'}).getText()
    market['price_VND'] = item.find('td', {'class':'text-right text-danger font-weight-bold'}).getText()
    market['date'] = item.find_all('td', {'class':'text-right'})[1].getText()
    market['city'] = item.find_all('a',)[1].getText()
    #market['city'] = item.find_all('a',{'class':'text-truncate align-middle'}).getText()


    markets.append(market)

shrimp_pricing = pd.DataFrame(markets)
shrimp_pricing
```

```python
shrimp_pricing['price_VND'] = shrimp_pricing['price_VND'].astype(float)
```

```python
shrimp_pricing['date_official'] = pd.to_datetime(shrimp_pricing['date'], format='%d-%m-%y')
shrimp_pricing
```

### 4.1 exporting prices for swine into a CSV file

```python
import datetime

current_date = datetime.datetime.now()
filename = str(current_date.day)+str(current_date.month)+str(current_date.year)
shrimp_pricing.to_csv(str('../Pig333/data/AQUAVND/AQUApricetoday'+filename+ '.csv'))
#pricing.to_csv('/Users/lstorino/PythonTests/scrapping/Pig333/data/pricetoday.csv')
```

```python
shrimp_pricing['price_USD'] = shrimp_pricing.price_VND.div(23.3).round(2)
shrimp_pricing
```

```python
aqua_scene = shrimp_pricing[['specie','date','price_USD']].style.hide_index()
aqua_scene
```
